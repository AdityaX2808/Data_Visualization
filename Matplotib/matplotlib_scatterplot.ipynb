<h1>Matplotlib Scatterplot Problems</h1>

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

1. Write a Python program to draw a scatter graph taking a random distribution in X and Y and
plotted against each other.

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y, color='blue')
plt.title("Scatter Plot with Random Distribution")
plt.show()
```

2. Write a Python program to draw a scatter plot with empty circles taking a random distribution
in X and Y and plotted against each other

```python
plt.scatter(x, y, edgecolors='blue', facecolors='none')
plt.title("Scatter Plot with Empty Circles")
plt.show()
```

3. Write a Python program to draw a scatter plot using random distributions to generate balls of
different sizes.

```python
sizes = np.random.rand(50) * 1000
plt.scatter(x, y, s=sizes, color='green', alpha=0.5)
plt.title("Scatter Plot with Different Sizes")
plt.show()
```

5. Write a Python program to draw a scatter plot for three different groups camparing weights
and heights.

```python
group1 = np.random.rand(10, 2) * 100
group2 = np.random.rand(10, 2) * 100 + 50
group3 = np.random.rand(10, 2) * 100 + 100

plt.scatter(group1[:, 0], group1[:, 1], color='blue', label='Group 1')
plt.scatter(group2[:, 0], group2[:, 1], color='green', label='Group 2')
plt.scatter(group3[:, 0], group3[:, 1], color='red', label='Group 3')
plt.xlabel("Weight")
plt.ylabel("Height")
plt.title("Scatter Plot for Three Groups")
plt.legend()
plt.show()
```
